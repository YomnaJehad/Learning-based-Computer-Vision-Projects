### . Load the images into arrays and resize them.

```python
# import necessariy libraries
from skimage.io import imread
from skimage.transform import resize

import numpy as np
from numpy import asarray
from matplotlib import pyplot
import os
```

```python
# define the required paths
base_path = 'EnglishImg/English/Img/GoodImg/Bmp/'
o_path = 'Sample051'
q_path = 'Sample053'
G_path = 'Sample017'
n8_path = 'Sample009'
paths = [o_path, q_path, G_path, n8_path]

# define a list for all images
images_list = [[], [], [], []]

for i in range (0,4):
    
    # extract a list of file names in each directory
    files_list = os.listdir( base_path+paths[i] )
    print('Numebr of images in',paths[i],': ' ,len(files_list))
    
    for item in files_list:
        image = imread(base_path+paths[i]+'/'+item)
        # resize the images to 32,32
        resized = resize(image, (32, 32)) 
                    
#         pyplot.imshow(resized)
#         pyplot.show()
        # append to the list as a numpy array
        images_list[i].append(asarray(resized))

# show an example
pyplot.imshow(images_list[0][0])
pyplot.show()
```

Before we begin, let's prepare some functions to make the process easier and more clean

```python
# define a function for training binary classifier  
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn import metrics

def applyBinaryClassifier (X_train, y_train, X_test, y_test, max_iter=10, tol=0.1, solver ='liblinear', penalty='l2', C=1):
    
    # define the classifier and train it
    bin_clf = LogisticRegression(max_iter=max_iter, tol=tol, solver=solver, penalty=penalty, C=C)
    bin_clf.fit(X_train, y_train)
    
    # print training and test accuracies
    print( "Train Accuracy: ", bin_clf.score(X_train, y_train) )
    print( "Test Accuracy: ", bin_clf.score(X_test, y_test) )
    print('-----------------------------')
    return bin_clf
```

```python
# define a function for getting binary classifier metrics like confusion matric and so on
def getBinaryMetrics (bin_clf, X, y):
    
    # plot confusion matrix
    plot_confusion_matrix(bin_clf, X, y)  
    plt.show()
    
    
    # calculate recall, percision and accuracy (accuracy-of course- should be same as the above result)
    y_pred = bin_clf.predict(X)
    TN, FP, FN, TP = confusion_matrix(y, y_pred).ravel()
    Accuracy = (TP + TN) / (TP + TN + FP + FN)
    Recall = TP / (TP + FN)
    Precision =  TP / (TP + FP)
    F1 = (2 * Precision * Recall) / (Precision + Recall)
    print(f"Model Accuracy: {Accuracy}, Model Recall: {Recall}, Model Precision: {Precision} F1Score:{F1:.2f} ")
    
    # plot roc curve
    
    
    metrics.plot_roc_curve(bin_clf, X, y)
    plt.plot([0,1],[0,1],'--')
    plt.show()
```

```python
# define a function to run the pipeline
def pipeline (X_train, y_train, X_test, y_test, max_iter=10, tol=0.1, solver ='liblinear', penalty='l2', C=1):
    
    bin_clf = applyBinaryClassifier (X_train, y_train, X_test, y_test,
                                     max_iter=max_iter, tol=tol, solver =solver, penalty=penalty, C=C)
    print("Train Metrics")
    getBinaryMetrics (bin_clf, X_train, y_train)
    print("_________________________________________\n")
    print("Test Metrics")
    getBinaryMetrics (bin_clf, X_test, y_test)
    print("_________________________________________\n")
```

## Result Explanation

I tried out various hyperparameters vs various data splitting techniques and the champion combination was: shuffle and stratify for the sets with quite high regularization penalty. <br> Because the classes were imbalanced o:148, q:54 and G:143 8:32. So the stratifying ensures that the train and test sets have approximately the same percentage of samples of each target class as the complete set which gives real results. Otherwise we couldn't trust this model if given more of the smaller class in the real world. <br>
The 'lbfgs' solver performed better than 'liblinear'. according to [this](https://stackoverflow.com/questions/38640109/logistic-regression-python-solvers-definitions)
, since the dataset is small, it performs better given enough iterations to converge. (This performance improvement only happened in the 1st classifier, the 2nd didn't show significant change)<br>
Also, increasing the penality strength through decreasing C to prevent overfitting as much as I could.<br>
Throughout these two models train and test accuracy metrics. It's obvious that the precision is high while the recall is low. Which means that the model hardly makes any false positives, but makes more false negatives (we don't really have an interest in one class over the other here). In short, it means that one class is barely classified while another class is classified much more often even if it's not true.
Which, again, is a result of the imbalanced dataset. <br>
Regarding the ROC, for the training set it shows good rates and a really good area under the curver. However for the test set, it's barely crossing the random classifier boundary, which means there's a bit of overfitting to the training set. <br>
I belive most of these problems can be solved by upsampling the smaller classes, and also by adding some augmented images.

```python
import pandas as pd
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import seaborn as sns

def plot_multiclass_roc(multi_clf, X, y, n_classes, figsize=(10, 7)):
    
    y_score = multi_clf.decision_function(X)

    # structures
    fpr = dict() # define false positive rate
    tpr = dict() # define true positive rate 
    roc_auc = dict() #define roc_auc

    # calculate dummies once for the true y values (This is important to complete the plot)
    y_dummies = pd.get_dummies(y, drop_first=False).values
    for i in range(n_classes):
        # calculate rates from the score value and it's corresponding true y value
        fpr[i], tpr[i], _ = roc_curve(y_dummies[:, i], y_score[:, i])
        # calculate areas 
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

##### Note: The previous function idea reference [here](https://stackoverflow.com/questions/45332410/roc-for-multiclass-classification). I took my time understanding it and editing it to fulfill my purpose

```python
# define a function for getting binary classifier metrics like confusion matric and so on
def getMultiMetrics (multi_clf, X, y):
    
    # plot confusion matrix
    plot_confusion_matrix(multi_clf, X, y)  
    plt.show()
    
    
    y_pred = multi_clf.predict(X)
    t0p0, t0p1, t0p2, t0p3, t1p0, t1p1, t1p2, t1p3, t2p0, t2p1, t2p2, t2p3, t3p0, t3p1, t3p2, t3p3 = confusion_matrix(y, y_pred).ravel()
    
    # Metrics for the class 0 "o"
    Recall_0 = t0p0 / (t0p0 + t0p1 + t0p2 + t0p3)
    Percision_0 =  t0p0 / (t0p0 + t1p0+ t2p0 + t3p0)
    print(f"Class o Recall Value: {Recall_0:.3f}, Precision: {Percision_0:.3f}")
    
    # Metrics for class 1 "q"
    Recall_1 = t1p1 / (t1p1 + t1p0 + t1p2 + t1p3)
    Percision_1 =  t1p1 / (t1p1 + t0p1+ t2p1 + t3p1)
    print(f"Class q Recall Value: {Recall_1:.3f}, Precision: {Percision_1:.3f}")
    
    # Metrics for class 2 "G"
    Recall_2 = t2p2 / (t2p2 + t2p1 + t2p0 + t2p3)
    Percision_2 =  t2p2 / (t2p2 + t1p2+ t0p2 + t3p2)
    print(f"Class G Recall Value: {Recall_2:.3f}, Precision: {Percision_2:.3f}")
    
    # Metrics for class 3 "8"
    Recall_3 = t3p3 / (t3p3 + t3p1 + t3p2 + t3p0)
    Percision_3 =  t3p3 / (t3p3 + t1p3+ t2p3 + t0p3)
    print(f"Class 8 Recall Value: {Recall_3:.3f}, Precision: {Percision_3:.3f}")
    
    Recall_avg = (Recall_0+Recall_1+Recall_2+Recall_3)/4
    Percision_avg = (Percision_0+Percision_1+Percision_2+Percision_3)/4
    print(f"Average Recall Value: {Recall_avg:.3f}, Average Precision Value: {Percision_avg:.3f}")
    
    F1Score = (2 * Percision_avg * Recall_avg) / (Percision_avg + Recall_avg)
    print(f"F1 Score: {F1Score:.3f}")
    
    Accuracy = (t0p0+t1p1+t2p2+t3p3)/ (t0p0+t0p1+t0p2+t0p3+t1p0+t1p1+t1p2+t1p3+t2p0+t2p1+t2p2+t2p3+t3p0+t3p1+t3p2+t3p3)
    print(f"Overall accuracy: {Accuracy:.3f}")
  
    plot_multiclass_roc(multi_clf, X, y, n_classes=4)
```

##### Note The Multiclass F1 Score and Accuracy 's formulas reference [here](https://arxiv.org/pdf/2008.05756.pdf#:~:text=Accuracy%20is%20one%20of%20the,computed%20from%20the%20confusion%20matrix.&text=The%20formula%20of%20the%20Accuracy,confusion%20matrix%20at%20the%20denominator.).

## Result Explanation

The model accuracy is barely 55%, also the precision recall values vary between high and low for each class, this is due to the imbalanced classes. This variation can tell us some information about each class prediction and misprediction scenarios.<br>
For example, in the test metrics, the letter "G" has higher recall and lower precision which means there are a lot of negative examples (none "G" classes) that could become false positives (falsely classified as "G"). Conversely, there are fewer positive (real "G") examples that could become false negatives (classified as non "G"). In short, the classifier predicts so many "G"s than there really is. This could be due to the fact that "G" has a high number of training examples compared to the other classes. The same case applies to class "o" which has the highest number of training examples.<br>
Regarding the classes "q" and "8", they're the opposite. Having the lowest training examples, and thus resulting in really low Recall and high precision. Which means, the model barely (if at all) classifies the inputs as either. by looking at the ROC the "8" class is almost a random classifiers. <br>
As mentioned before, I believe the solution to most of these problems is either getting more images of the lower classes or using resampling methods to balance the dataset. Another thing, in order to fix the gap between the train and test accuracy, more augmented images and further regularization techniques could be used. <br>
